```python
driver=webdriver.Chrome(r"C:\web driver\chromedriver_win32\chromedriver.exe")
driver=webdriver.Chrome("Chromedriver.exe")
time.sleep(1)

driver.get("https://www.google.com/maps")
```

```python
b=str(input("Enter a Place = "))
search_field_designation=driver.find_element_by_xpath("/html/body/div[3]/div[9]/div[3]/div[1]/div[1]/div[1]/div[2]/form/div[2]/div[3]/div/input[1]")
search_field_designation.send_keys(b)

search_button=driver.find_element_by_xpath("/html/body/div[3]/div[9]/div[3]/div[1]/div[1]/div[1]/div[2]/div[1]/button")
search_button.click()
```

```python
c=driver.current_url
la=[]
lo=[]
la.clear()
lo.clear()
k=0
for i in c:
    k=k+1
    if i=='@':
        for j in range(k,k+10,1):
            la.append(c[j])
        for l in range(k+11,k+21,1):
            lo.append(c[l])
    else:
        continue

print("Latitude = ","".join(map(str, la)))
print("Longitude = ","".join(map(str, lo)))
```
